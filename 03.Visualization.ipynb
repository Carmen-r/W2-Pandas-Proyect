### Librerías

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

#### Importamos el dataset

```python
df = pd.read_csv("dataset/midatasetslimpio.csv/attack_cleanhipotesis.csv")
df2 = pd.read_csv("dataset/midatasetslimpio.csv/attack_cleanhipotesis2.csv")
df3 = pd.read_csv("dataset/midatasetslimpio.csv/attack_cleanhipotesis3.csv")
```

#### 1ª Hipótesis

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
zone = sns.countplot(x=df.Months, hue=df.Hemispheres, order=["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"])
ax.set_title("Months and Hemispheres")
```

```python
zone.figure.savefig("images/1sth.svg")
```

#### 2ª Hipótesis

```python
df2
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
country = sns.countplot(x=df2.Country, palette="Blues")
ax.set_title("Attacks by Country")
```

```python
country.figure.savefig("images/2nd.svg")
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
activity = sns.countplot(x=df3.Activities, palette="Greens")
ax.set_title("Attacks by Activity")
```

### Limpieza preliminar de los datos

```python
#First, is important to make a copy and delete this values in the copy.
df_ = df.copy()
df_
```

```python
#In this case, there are a lot of values with no information. 
df_sum = df_.isna().sum()
df_sum
```

```python
#We are going to delete the columns that cointain more than 20.000 null values because is not part of the study.
drop_cols = list(df_sum[df_sum > 20000].index)
df_drop = df_.drop(drop_cols, axis=1)
df_drop
```

```python
#We have to eliminate those columns that will not be of interest to us in the study.
df__ = df_drop.drop(['Year', 'Type','Sex ','Fatal (Y/N)', 'Injury', 'Case Number', 'Area', 'Location', 'original order', 'Name', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2'], axis = 1)
df__
```

`We have already eliminated all those columns that are not the object of study, now we have to eliminate all those rows with null values`

```python
df_clean = df__.dropna(axis = 0, how = "all")
df_clean
```

## SUMMARY

The first project at Ironhack has been about shark attacks around the world. For this, we downloaded a kaggle file with all the data collected in a table. The first step has been to perform a preliminary cleaning to keep those columns that are the object of study. Then, we performed a deep cleaning since not all the words were written in the same way and were useful for the project. Secondly, once we have the tables we have created graphs for their visualization.

### First Conclusion

`- The highest number of attacks occurred in the Northern Hemisphere in summer.`

In this graph we can observe the peaks of both Hemispheres are in its summer season. In contrast, the lows are in its winter. Thus, we can conclude the hotest is the water, more people are attacked. Also, the chances of being attacked are higher in summer as more people go to the beach

### Second Conclusion

`- The highest number of attacks occurred in Australia.`

We can see above the most attacked country is USA. Thereby, the formulated hypothesis is negative due to Australia is the second more attacked country.

### Third Conclusion

`- Number of attacks depending on the water activity you are doing.`

This graph shows the chances of sufering an attack are higher if you are surfing rather than fishing or bathing.
